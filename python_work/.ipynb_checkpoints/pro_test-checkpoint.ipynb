```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 

from matplotlib import rc, font_manager
font_name=font_manager.FontProperties(fname='c:/Windows/Fonts/batang.ttc').get_name()
rc('font',family=font_name)

#그림의 크기
plt.rcParams['figure.figsize']=(12,9)
```

```python
fire=pd.read_csv('data/발화요인별_화재발생_20211209120155.csv',sep=',',header=1,engine='python',encoding='cp949')
del fire['시점']
fire.rename(columns={'소계.1':'자연적요인'},inplace=True)
fire.rename(columns={'소계.2':'발화요인 미상'},inplace=True)
fire.rename(columns={'소방서별(1)':'소방서별'},inplace=True)
fire.drop(fire.loc[fire['소방서별']=='합계'].index, inplace=True)

fire['가스누출'] = fire['가스누출'].apply(lambda x: 0 if x=='-' else x)
fire['화학적요인'] = fire['화학적요인'].apply(lambda x: 0 if x=='-' else x)
fire['교통사고'] = fire['교통사고'].apply(lambda x: 0 if x=='-' else x)
fire['기타'] = fire['기타'].apply(lambda x: 0 if x=='-' else x)
fire['자연적요인'] = fire['자연적요인'].apply(lambda x: 0 if x=='-' else x)
fire['방화'] = fire['방화'].apply(lambda x: 0 if x=='-' else x)
fire['방화의심'] = fire['방화의심'].apply(lambda x: 0 if x=='-' else x)

fire=fire.astype({'가스누출':'int'})
fire=fire.astype({'화학적요인':'int'})
fire=fire.astype({'교통사고':'int'})
fire=fire.astype({'기타':'int'})
fire=fire.astype({'자연적요인':'int'})
fire=fire.astype({'방화':'int'})
fire=fire.astype({'방화의심':'int'})

fire=fire.groupby('소방서별').sum()
fire.reset_index(level=['소방서별'], inplace = True)
fire=fire.sort_values('소방서별')

print(fire.shape)

fire
```
