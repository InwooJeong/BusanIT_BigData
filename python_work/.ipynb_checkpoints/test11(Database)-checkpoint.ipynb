```python
import sqlite3
print(sqlite3.version)
print(sqlite3.sqlite_version)
```

```python
def create_table():
    conn=sqlite3.connect('my_book.db')
    cursor=conn.cursor()
    sql="""CREATE TABLE if not exists books(
    title text,
    published_date text,
    publisher text,
    pages integer
    )"""
    cursor.execute(sql)
    conn.commit()
    conn.close()
create_table()
```

```python
def insert_books(data):
    conn=sqlite3.connect('my_book.db')
    cursor=conn.cursor()
    #sql="""INSERT INTO books 
    #      values('java','2021-01-12','길벗',500)"""
    sql="INSERT INTO books values(?,?,?,?)"
    cursor.execute(sql,data)
    conn.commit()
    conn.close()
```

```python
data=('python','2011-01-12','한빛',450)
insert_books(data)
```

```python
def insert_many_books(data):
    conn=sqlite3.connect('my_book.db')
    cursor=conn.cursor()
    #sql="""INSERT INTO books 
    #      values('java','2021-01-12','길벗',500)"""
    sql="INSERT INTO books values(?,?,?,?)"
    cursor.executemany(sql,data)
    conn.commit()
    conn.close()
```

```python
items=[('Big data','2021-05-05','영진',550),('android','2022-01-01','한빛',600)]
insert_many_books(items)
```

```python
def all_books():
    conn=sqlite3.connect('my_book.db')
    cursor=conn.cursor()
    sql="""SELECT  *
           FROM    books"""
    cursor.execute(sql)
    books=cursor.fetchall()
    print(books)
    print(type(books))
    
    for book in books:
        print(book)
        print(type(book))
    conn.close()
```

```python
all_books()
```

```python
def some_books(number):
    conn=sqlite3.connect('my_book.db')
    cursor=conn.cursor()
    sql="""SELECT  *
           FROM    books"""
    cursor.execute(sql)
    books=cursor.fetchmany(number)
    for book in books:
        print(book)
    conn.close()
```

```python
some_books(3)
```

```python
def one_book():
    conn=sqlite3.connect('my_book.db')
    cursor=conn.cursor()
    sql="""SELECT  *
           FROM    books"""
    cursor.execute(sql)
    books=cursor.fetchone()
    for book in books:
        print(book)
    conn.close()
```

```python
one_book()
```

```python
def select_books(data):
    conn=sqlite3.connect('my_book.db')
    cursor=conn.cursor()
    sql="""SELECT   *
           FROM     books
           WHERE    pages>? or publisher=?
           ORDER BY title"""
    cursor.execute(sql,data)
    books=cursor.fetchall()
    for book in books:
        print(book)
    conn.close()
```

```python
data=(500,'한빛')
select_books(data)
```

```python
def update_book(data):
    conn=sqlite3.connect('my_book.db')
    cursor=conn.cursor()
    sql="UPDATE books SET pages=? WHERE title=?"
    cursor.execute(sql,data)
    conn.commit()
    conn.close()
```

```python
update_book((700,'android'))
all_books()
```
