```python
#산점도 : 서로 다른 독립변수 x1,x2의 관계를 파악
#normal(mu,sigma,size(rows,cols)) mu중앙값, sigma표준편차
x1=np.random.normal(1,1,size=(100,1))
x2=np.random.normal(-2,4,size=(100,1))
#concatenate:열방향 연결, axis=1 가로, 0 세로
x=np.concatenate((x1,x2),axis=1)
print(x[:5])
df3=pd.DataFrame(x,columns=['x1','x2'])
df3.head()
```

```python
plt.scatter(df3.x1,df3.x2)
```

```python
import csv
from datetime import datetime
filename='data/temperature_2014.csv'
with open(filename) as f:
    reader=csv.reader(f)
    header_row=next(reader)
    dates,highs,lows=[],[],[]
    for row in reader:
        try:
            #strftime : 날짜 -> 문자
            #strptime : 문자 -> 날짜
            current_date=datetime.strptime(row[0],'%Y-%m-%d')
            high=int(row[1])
            low=int(row[3])
        except ValueError:
            print(current_date,'missing date')
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)
            
flg=plt.figure(dpi=128,figsize=(10,6))
plt.plot(dates,highs,c='red',alpha=0.5)
plt.plot(dates,lows,c='blue',alpha=0.5)
plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.1)

plt.title('온도 통계',fontsize=20)
plt.xlabel('기간',fontsize=16)
plt.ylabel('온도',fontsize=16)
plt.tick_params(axis='both',labelsize=16)

plt.savefig('data/temperature_plot.png', dpi=400, bbox_inches='tight')
plt.show()
```
