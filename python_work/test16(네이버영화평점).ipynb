```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
def getLinks(page):
    html=urlopen("https://movie.naver.com/movie/point/af/list.naver?&page="+str(page))
    #print(html)
    obj=bs(html,'html.parser')
    all_text=obj.select('td.title')
    data_list=[]
    for s1 in all_text:
        text=s1.get_text()
        textarr=text.replace('신고','').replace('\t','').replace('\n\n\n\n','').split("\n\n")
        textarr[0]=textarr[0].strip('\n')
        textarr[1]=int(textarr[1][textarr[1].find('중')+1:])
        if len(textarr)<3:
            textarr.append('-')
        data_list.append(tuple(textarr))
    return data_list
```

```python
getLinks(1)
```

```python
html=urlopen("https://movie.naver.com/movie/point/af/list.naver")
obj=bs(html,'html.parser')
title_list=[]
point_list=[]
review_list=[]
movie_reviews=[]
titles=obj.select('td.title>a')
for i in range(len(titles)):
    if i%2==0:
        title=titles[i].string
        title_list.append(title)
#print(title_list)
points=obj.select('td.title>div>em')
for point in points:
    point_list.append(int(point.string))
#print(point_list)
reviews=obj.select('td.title')
for review in reviews:
    text_arr=review.text.replace('신고','').replace('\t','').replace('\n\n\n\n','').split("\n\n")
    if len(text_arr)<3:
        review_list.append('-')
    else:
        review_list.append(text_arr[2])
        
#print(review_list)

for i in range(10):
    movie_reviews.append(tuple([title_list[i],point_list[i],review_list[i]]))
print(movie_reviews)
```
