```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
def func(url):
     res=req.urlopen(url)
     soup=BeautifulSoup(res,'html.parser')
     content=soup.select('div.mw-parser-output p')
     if len(content)>0:
         print(content[0].text)
```
