```python
from collections import Counter
from konlpy.tag import Okt
import pytagcloud
```

```python
f=open('data/data.txt',encoding='utf-8')
data=f.read()
print(data)
```

```python
npl=Okt()
nouns=npl.nouns(data)
print(nouns)
```

```python
count=Counter(nouns)
print(nouns)
```

```python
tag2=count.most_common(20)
taglist=pytagcloud.make_tags(tag2,maxsize=80)
print(taglist)
f.close()
```

```python
pytagcloud.create_tag_image(taglist,'data/wordcloud.jpg',size=(900,600),fontname='korean',rectangular=False)
```

```python
wordInfo={}
for tag, count in count.most_common(50):
    if(len(str(tag))>1):
        wordInfo[tag]=count
        print('%s:%d'%(tag,count))
print(wordInfo)
```
