```python
titanic=sns.load_dataset('titanic')
print(titanic.shape)
titanic.head()
```

```python
#tips 데이터 셋 load
tips=sns.load_dataset('tips')
print(tips.shape)
tips.head()
```

# 1. countplot()
- 항목의 갯수를 세어줌
- 알아서 column을 구성하고 있는 value들을 구분하여 보여줌

```python
sns.countplot(x='class',hue='who',data=titanic)
plt.show()
```

```python
sns.countplot(x='day',hue='sex',data=tips)
plt.show()
```

```python
sns.countplot(y='day',hue='sex',data=tips)
plt.show()
```

```python
sns.countplot(y='class',hue='who',data=titanic)
plt.show()
```

```python
sns.countplot(x='class',hue='who',palette='Accent',data=titanic)
#palette 값 입력 오류시 사용가능한 palette명 표시
plt.show()
```

```python
x=np.random.randn(100)
x
```

```python
sns.distplot(x)
```

```python
#rugplot
#rug는 데이터 위치를 x축 위에 작은 선분rug로 나타내, 데이터들의 위치 및 분포를 보여준다.
sns.distplot(x,rug=True,hist=False,kde=True)
```

```python
sns.distplot(x,vertical=True)
```

```python
#컬러 변경 r,g,b,c,m,y 가능
sns.distplot(x,color='y')
```
