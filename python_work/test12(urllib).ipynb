```python
import urllib.request
from urllib.request import urlopen
```

```python
#res=urllib.request.urlopen("http://www.naver.com")
res=urlopen("http://www.naver.com")
print(type(res))
print(res.status)
print(res.version)
print(res.msg)
res_headers=res.getheaders()
print("="*20)
for s in res_headers:
    print(s)
#print(res.read())
```

```python
res1=urlopen("https://movie.naver.com/movie/point/af/list.naver")
encoding=res1.info().get_content_charset()
print("encoding:",encoding)
print(res1.read().decode('utf-8'))
```

```python
from urllib.error import HTTPError
from urllib.error import URLError
try:
    html=urlopen("http://java10.com")
except HTTPError as e:
    print("HTTPError 입니다.")
except URLError as e:
    print("존재하지 않는 사이트입니다.")
else:
    print(html.read())
```

```python
from urllib.request import urlopen
from urllib.parse import urlencode
```

```python
API="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
# 지역번호 : 전국 108, 서울/경기 109, 강원 105, 충북 131 충남 133,
# 전북 146, 전남 156, 경북 143, 경남 159, 제주 184
values={'stnId':184}
params=urllib.parse.urlencode(values)
#요청 전용 url 생성
url=API+"?"+params

print("url=",url)
#다운로드
data=urllib.request.urlopen(url).read()
text=data.decode('utf-8')
print(text)
```
